```python
unique_marital=bank_df['marital'].unique()
for i in unique_marital:
    con=bank_df['marital']==i
    val=len(bank_df[con])
    print(f"{i}:{val}")
```

```python
unique_marital=bank_df['marital'].unique()
count_list=[]
for i in unique_marital:
    con=bank_df['marital']==i
    val=len(bank_df[con])
    count_list.append(val)
```

```python
marital_df=pd.DataFrame(zip(unique_marital,count_list),
                    columns=['marital','count'])
marital_df
```

```python
plt.figure(figsize=(8,5))
plt.bar('marital','count',
        data=marital_df)
plt.title("Bar Chart")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("Barchart.jpeg")
plt.show()
```

```python
plt.figure(figsize=(15,10))
mdf=bank_df['marital'].value_counts()
keys=mdf.keys()
sns.countplot(data=bank_df,
              x='marital',
               order=keys)
plt.title("Bar Chart")
plt.xlabel("marital")
plt.ylabel("count")
plt.show()
```

```python
import matplotlib.pyplot as plt
mdf=bank_df['marital'].value_counts()
keys=mdf.keys()
values=mdf.values
keys,values
```

```python
plt.pie(values,
        explode=[0.1,0.1,0.1,0.1],
        labels=keys,
        autopct="%0.2f%%",
        startangle=180,
        radius=1)
plt.show()
```
