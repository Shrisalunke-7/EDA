- we have some transformation Techni are there

**Step1**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
exp_data=np.random.exponential (size=10000)
exp_data
```

```python
plt.hist(exp_data,bins=20,label='Exponential')
plt.show()
```

```python
norm_data=np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Exponential')
plt.show()
```

**The goal is exp_data to normal distribution**
- here bins=interval


**Step4**
- Log transformation
- log trans means performaing logarithmic operations on original data
- it is ans approach to convert into normal distribution
- log means natural logarithmic base=e

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
```

**Reciprocal Transformations**
- Reciprocal fails when values are has very less
- for eg. x=0 then and 1/0 fail

```python
reciprocal_data=np.reciprocal(exp_data)
plt.hist(reciprocal_data,bins=20,label='Reciprocal')
plt.show()
```

**Square Root Transformation**

```python
square_root_data=np.sqrt(exp_data)
plt.hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,3,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.subplot(1,3,3).hist(square_root_data,bins=20,label='Square Root Data')
plt.legend()
plt.show()
```
