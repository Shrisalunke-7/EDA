```python
#Read data
visa_df=pd.read_csv(r'C:\Users\raj\Documents\EDA\Visadataset.csv')
```

**Advantages**
- In ML we have one important property  is independent of variables
- ML models expects variables are independent of each other
- One hot encoder creating column, the new column is independent to each other
- indepent variable means both variable haave**90** degree phase shift
- 90 degree phase shift means also callled as perpendicular each other
- perpendicular means orthogonal each other

**Disadvantages**

- one hot encoder create new columns = number of unique labels
- If no. of unique labels increase, new columns also increase
- the more columns leads datra is more 
- then data is more hte compexity will increase
- complexity increases more computation power required 
- also time aslo increase
- the entire process names as **curse of dimensionality**
